# Heat Simulations

This notebook shows solution behavior for a forced cubic heat equation.

Problem statement:

$$
\begin{align*}
    \begin{aligned}
    \frac{\partial q}{\partial t}
    &= \mu \frac{\partial^{2} q}{\partial x^{2}} - q(x,t)^{3} + f(x, t),
    && x\in(0, L), t\in(0, T],
    \\
    q(0, t) &= \alpha,
    \quad
    q(L, t) = \beta,
    && t\in(0, T],
    \\
    q(x, 0) &= q_{0}(x),
    && x\in [0, L].
    \end{aligned}
\end{align*}
$$

The classes used here solve this problem for different choices of forcing function $f(x, t)$.
The constructors take in the spatial domain, the boundary conditions $\alpha$ and $\beta$, and the diffusion constant $\mu$.

```python
import numpy as np
import matplotlib.pyplot as plt

import mpl_config
import pde_models as pdes
```

```python
alpha = 0
beta = 1
mu = 5e-3
x = np.linspace(0, 1, 500)
t = np.linspace(0, 2, 500)

heatsolver_cubic = pdes.CubicHeatBimodal(x, alpha, beta, diffusion=mu)
initial_conditions = heatsolver_cubic.initial_conditions(x, alpha, beta)
solution = heatsolver_cubic.solve(initial_conditions, t)
```

```python
fig, axes = heatsolver_cubic.plot_spacetime(t, solution, nlines=12)
axes[0].legend([])
fig.suptitle("")
axes[0].set_title("profiles in space")
axes[1].set_title("space-time behavior")
plt.show()
```
