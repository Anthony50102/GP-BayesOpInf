# Epidemiological Model Simulations

```python
import numpy as np
import matplotlib.pyplot as plt

import mpl_config
import ode_models as odes
```

## SEIRD Model

```python
N = 1.0
beta = 0.25
delta = 0.1
gamma = 0.1
alpha = 0.05
rho = 0.05
E0, I0, R0, D0 = 0.005, 0.001, 0.0, 0.0
S0 = N - E0 - I0 - R0 - D0
t = np.linspace(0.0, 200, 201)

# Set up and solve the model.
seird = odes.SEIRD([N, beta, gamma, delta, alpha, rho])
solution = seird.solve(initial_conditions=[S0, E0, I0, R0, D0], timepoints=t)
```

```python
# Plot the solution without noise.
seird.plot(t, solution, ls="-")
plt.show()
```

```python
# Plot the solution with noise.
fig, ax = seird.plot(t, seird.noise(solution, 0.1))
ax.set_ylim(top=1.4)
ax.legend(loc="upper center", ncol=5)
plt.show()
```
